# Importing price data from Yahoo Finance and exporting to .csv

```python
# calling necessary libraries
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

# Using shift() method to calculate absolute and relative change in stock price over previous period (last week to this week, yesterday to today etc.)

```python
close.head()
```

```python
# copying AAPL closing data to dataframe
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
# To calculate price increase from last day, we need to shift all rows down by 1
# shift method shifts the index by a desired number of periods
# periods can can take any integer value, -1, 3, 5..
aapl.shift(periods = 1)
```

```python
# creating new column to save the shifted data
aapl['lag1'] = aapl.shift(periods = 1)
```

```python
aapl.head()
```

```python
# Calculating daily increase in Prices by subtracting AAPL column by lag1
aapl.AAPL.sub(aapl.lag1)
```

```python
# Creating a new Diff column to save the increase in prices
aapl['Diff'] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl.head()
```

```python
# Creating a new column Percentage change
# dividing values in AAPL column by lag1 column, subtracting by 1 and then multiplying by 100
aapl['Pct. Change'] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl.head()
```

# Calculating Absolute and Relative changes in time series using diff() and pct_change() methods

```python
aapl.head()
```

```python
# Calculating absolute difference with diff() method
aapl.AAPL.diff(periods = 1)
```

```python
aapl['Diff2'] = aapl.AAPL.diff(periods = 1)
```

```python
aapl.head(10)
```

```python
# Checking whether the difference results from shift() method in Diff column and from diff() method in Diff2 column are same
# We use equals() to check that
aapl.Diff.equals(aapl.Diff2)
```

```python
# Using pct_change() to calculate percentage change
aapl['pct_change2'] = aapl.AAPL.pct_change(periods = 1).mul(100)
```

```python
aapl.head()
```

```python
# Resampling data into monthly values to calculate monthly returns
aapl.AAPL.resample('M').last()
# gives out the last value of the month
```

```python
# Since, stocks are traded only on weekdays, we can use BM(Business Month) instead of M 
aapl.AAPL.resample('BM').last()
```

```python
# Applying pct_change() to calculate monthly returns
aapl.AAPL.resample('BM').last().pct_change(periods =1).mul(100)
```

# Measuring stock performance with MEAN Return and STD of Returns

```python
# importing necessary libraries
import numpy as np
```

```python
# copying AAPL closing prices to a new dataframe
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
# Calculating percentage change and dropping the NaN value using dropna()
# by default period in pct_change is 1
aapl.pct_change().dropna()
```

```python
# Saving the dataframe as daily price reuturns using variable ret
ret = aapl.pct_change().dropna()
```

```python
ret.head()
```

```python
# Checking ret information
ret.info()
```

```python
# Visualising returns using a histogram
ret.plot(kind = 'hist', figsize = (12,8), bins = 100)
plt.show()
```

```python
# Mean and standard deviation(sqrt. variance) can be used to summarise these returns
# Risk of the investement is measured as variability or volatility of returns; measured as standard deviation of returns
# Investors prefer high returns and no volatility
```

```python
# Calculating daily mean return using mean()
daily_mean_return = ret.mean()
daily_mean_return
```

```python
# Calculating daily variance of returns
var_daily_Returns = ret.var()
var_daily_Returns
```

```python
# calculating standard deviation of daily returns
std_daily_Returns = np.sqrt(var_daily_Returns)
std_daily_Returns
```

```python
# can also calculate standard deviation of returns using std()
ret.std()
```

```python
# Annualising daily mean return by multiplying with 252, which is the avg. number of trading days in an year
ann_mean_Return = ret.mean() * 252
ann_mean_Return
```

```python
# Annualising variance and standard deviation similarly
ann_var_Returns = ret.var() * 252
ann_var_Returns
```

```python
# square root of 252 also needs to be taken when calculating st dev
ann_std_Returns = ret.std() * np.sqrt(252)
ann_std_Returns
```

```python
ret.head()
```

```python
# Observing the descriptive statistics of our return dataframe. Transposing the matrix using .T
# to select only mean and std columns we use loc
ret.describe().T.loc[:, ['mean','std']]
```

```python
# Saving the mean and std as summary dataframe
summary = ret.describe().T.loc[:,['mean','std']]
summary
```

```python
# Annualising the summary statistics
summary['mean'] = summary['mean'] * 252
summary['std'] = summary['std'] * np.sqrt(252)
```

```python
# Annualised  mean return and annualised standard deviation
summary
```

```python
# Creating scatter plot on our summary dataframe, and labelling each data point with the ticker symbol using for loop
summary.plot.scatter(x = 'std', y = 'mean', figsize = (12,8), s = 50, fontsize = 15)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,'std']+0.002, summary.loc[i, 'mean']+0.002), size = 15)
plt.xlabel('ann. Risk(std)', fontsize = 15)
plt.ylabel('ann. Return', fontsize = 15)
plt.title('Risk/Return', fontsize = 20)
plt.show()
```

# Simple Returns vs. Log Returns

```python
# Creating a dataframe with annual stock prices 
df = pd.DataFrame(index = [2016,2017,2018], data = [100,50,95], columns = ['Price'])
```

```python
df
```

```python
# Calculating simple returns
simple_returns = df.pct_change().dropna()
simple_returns
```

```python
# Calculating Mean of simple returns
simple_returns.mean()
```

```python
# Average of simple returns can be misleading, and does not give a clear picture of how much return was achieved.
# That is why we use log returns
```

```python
df
```

```python
# Calculating log returns
np.log(df / df.shift(1))
```

```python
# Creating a log_returns dataframe
log_returns = np.log(df / df.shift(1)).dropna()
```

```python
log_returns
```

```python
# Calculating mean of log_returns
log_returns.mean()
```

```python
# Mean log returns give a more accurate information of how much we can expect to make or lose on an investment
```

```python
# We can calculate current stock price using mean log returns if we have starting price and number of periods gone past
# This is impossible to do with simple returns
100 * np.exp(2 * log_returns.mean())

# Here 100 is the starting stock price, no. of time periods is 2
```
