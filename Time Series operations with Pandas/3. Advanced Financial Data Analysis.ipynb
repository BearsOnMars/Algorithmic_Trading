# Importing Financial Data from Excel

```python
import pandas as pd
```

```python
# read_excel methods works in the same way as read_csv method
pd.read_excel('SP500.xls')
```

```python
# Checking info to see the format of date column
pd.read_excel('SP500.xls').info()
```

```python
# but to be on the safe side, we can still Parse Dates and convert Date column into Index column
pd.read_excel('SP500.xls', parse_dates = ['Date'], index_col = 'Date')
```

```python
# We can choose which columns to import from excel file using a parameter usecols in pd.read_excel
# Here we are choosing to import columns A to E. By default pandas import all columns
# Suppose we need columns A till C, E and F to G, we can use usecols = 'A:C, E, F:G'
pd.read_excel('SP500.xls', parse_dates = ['Date'], index_col = 'Date', usecols = 'A:E')
```

```python
# We can specify which sheets to import from Excel. By default read_excel imports first sheet
pd.read_excel('SP500.xls', sheet_name = 'SP500')
```

```python
# We can also specify sheet number instead of its name. Sheet number index start with 0
pd.read_excel('SP500.xls', sheet_name = 1)
```

```python
# Importing and saving dataframe as SP500, and importing columns A to E
SP500 = pd.read_excel('SP500.xls', parse_dates = ['Date'], index_col = 'Date', usecols = 'A:E')
```

# Calculating Rolling Statistics with rolling()
Simple Moving Averages (SMA)

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# Importing the csv file
SP500 = pd.read_csv('SP500_Data_1971-2018.csv', parse_dates = ['Date'], index_col = 'Date')
```

```python
SP500.head()
```

```python
SP500.tail()
```

```python
SP500.info()
```

```python
# Choosing only close prices to work with and overwriting our dataframe SP500 using to_frame()
SP500 = SP500.Close.to_frame()
```

```python
SP500.head()
```

```python
# Plotting the prices
SP500.plot(figsize = (12,8), fontsize = 15)
plt.legend(loc = 'upper left', fontsize = 15)
plt.show()
```

```python
# Selecting the dates we want to focus on and analyse
SP500 = SP500.loc['2008-12-31':'2018-12-31'].copy()
```

```python
# Using rolling() to calculate rolling statistics. Choosing period as 10 here
SP500.rolling(window = 10)
```

```python
type(SP500.rolling(window = 10))
```

```python
SP500.head(15)
```

```python
# Creating a rolling average of 10 period aka Simple Moving Average (SMA)
SP500.rolling(window = 10).mean()
```

```python
# We can also calculate a rolling median(), max(), min().. 
SP500.rolling(window = 10).median()
```

```python
# by default the parameter for window is the same as that of min_periods
SP500.rolling(window = 10, min_periods = 5).mean()
```

# Exponentially-weighted Moving Averages (EWMA)

```python
# importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# Importing data from csv file and saving it to a dataframe
SP500 = pd.read_csv('SP500_Data_1971-2018.csv', parse_dates = ['Date'], index_col = 'Date', usecols = ['Date', 'Close'])
```

```python
SP500.head()
```

```python
# Choosing the dates we want to analyse, and overwriting SP500 variable dataframe
SP500 = SP500.loc['2008-12-31':'2018-12-31'].copy()
```

```python
SP500.Close.rolling(window = 10).mean()
```

```python
# Calculating Exponentially Weighted Moving Average using ewm(). Shift+Tab to check documentation
SP500.Close.ewm(span = 10, min_periods = 10).mean()
```

```python
# Adding SMA100 and EMA100 columns to SP500 Dataframe
SP500['SMA100'] = SP500.Close.rolling(window = 100).mean()
SP500['EMA100'] = SP500.Close.ewm(span = 100, min_periods = 100).mean()
```

```python
SP500
```

```python
# Plotting SMA100 and EMA100 moving averages to see the difference between two
SP500.iloc[:,-2:].plot(figsize = (15,10), fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
# It can be observed that EWMA is adapting faster to recent price changes than SMA
```

# Merging Time Series

```python
# Importing data from a csv file
stocks = pd.read_csv('stocks.csv', header = [0,1], index_col=[0], parse_dates = [0]).Close
```

```python
stocks.head()
```

```python
# Choosing a particular time period for AAPL and saving to a new dataframe variable
aapl = stocks.loc['2010-01-01':'2014-12-31', 'AAPL'].to_frame()
aapl.head()
```

```python
# Choosing a separate dataframe with Boeing prices with different period
ba = stocks.loc['2012-01-01':'2016-12-31', 'BA'].to_frame()
ba.head()
```

```python
# Merging BA column into the aapl dataframe
aapl['BA'] = ba.BA
```

```python
aapl.head()
```

```python
aapl.tail()
```

```python
# We can see a lot of missing values at the top because both AAPL and BA have different periods
# But pandas is smartly able to align the rest of the values with proper timestamps
```

```python
# Removing the empty BA rows using dropna()
aapl.dropna()
```

```python
# We can also use the reindex method to align these dissimilar dataframes AAPL and BA to get the similar output for BA
ba.reindex(aapl.index).dropna()
```

```python
# Incase we have a third time series with only friday closing prices, we can merge it as well
# Creating a new dataframe with DIS friday closing prices
dis = stocks.loc['2010-01-01':'2016-12-31', 'DIS'].resample('W-Fri').last().to_frame()
```

```python
aapl.head()
```

```python
# Creating an additional column in aapl dataframe for this new series
aapl['DIS'] = dis.DIS
```

```python
aapl.head(10)
```

```python
# We can reindex the DIS dataframe to show only those dates for which DIS closing price is available
dis.reindex(aapl.index)
```

```python
# Merging the two time series. It makes sense to align the data to the time series with lower frequency, in this case DIS
dis['AAPL'] = aapl.AAPL
```

```python
dis.head(10)
```

```python
# We can see that AAPL prices are now corresponding to the weekly friday close prices of DIS, with other values expunged
```
